BIG SALES DATA

```python
#Step 1: import library
import pandas as pd
```

```python
#Step 2: import data
big_sales = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Big%20Sales%20Data.csv')
```

```python
big_sales.head()
```

```python
# @title Item_Visibility

from matplotlib import pyplot as plt
big_sales['Item_Visibility'].plot(kind='line', figsize=(8, 4), title='Item_Visibility')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
big_sales.info()
```

```python
big_sales.describe()
```

```python
#step 3: define (y) and features (x)
big_sales.columns
```

```python
y = big_sales['Item_Outlet_Sales']
x = big_sales[['Item_Identifier', 'Item_Weight', 'Item_Fat_Content', 'Item_Visibility',
       'Item_Type', 'Item_MRP', 'Outlet_Identifier',
       'Outlet_Establishment_Year', 'Outlet_Size', 'Outlet_Location_Type',
       'Outlet_Type']]
```

```python
# Step 4 : train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.7, random_state=2529)
x_train

```

```python
# Step 5 : select model
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
# step 6 : train model
# Encode categorical features
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
```

```python
# Import SimpleImputer
le = LabelEncoder()
imputer = SimpleImputer(strategy='most_frequent')
```

```python
# Create an imputer object
for col in x_train.columns:
    if x_train[col].dtype == 'object':
        x_train[col] = le.fit_transform(x_train[col])
        x_test[col] = le.transform(x_test[col])
```

```python
# Impute missing values in numerical columns
x_train = imputer.fit_transform(x_train)
x_test = imputer.transform(x_test)

```

```python
# Convert 'y_train' to categorical (choose one method)
# 1. If ordinal:
y_train = pd.cut(y_train, bins=3, labels=False)
```

```python
# 2. If nominal and you know the threshold:
# threshold = 0
# y_train = (y_train > threshold).astype(int)
```

```python
# After converting 'y_train', refit the model:
model.fit(x_train, y_train)
```

```python
# step 7 : prediction
y_pred = model.predict(x_test)
x_test
```

```python
# step 8 : accuracy
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test,y_pred)
```

Explanation: Using RandomForest model the accuracy is 99.99%
